## Understanding the Mathematics
### Linear Equation
In a simple linear regression model, we predict the dependent variable $y$ as a linear function of the independent variable $x$. Assuming one independent variable, the model can be represented as:

$$ y = mx + c $$

where:
- $y$ is the dependent variable (output)
- $x$ is the independent variable (input)
- $m$ is the slope of the line (also known as the coefficient)
- $c$ is the y-intercept

In the context of machine learning, we usually write this equation as:

$$ y_i = \theta_0 + \theta_ix_i $$

where:
- $y_i$ is the predicted value (output)
- $x_i$ is the independent variable (input)
- $\theta_0$ is the y-intercept/constant term
- $\theta_i$ is the slope of the line with respect to the $i^{th}$ independent variable

### Error function and Minimization
The goal in linear regression is to find a hyperplane (or a line in 2D) that best fits our data. To do this, we first need to define what we mean by "best fit". For this, we define an error (or loss) function (in the context of machine learning referred to as cost function), which measures the difference between the predicted and the actual output. The most common way to define the "best fit line" is by minimizing the mean squared error (MSE). The MSE is defined as:
$$ MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Where:
- $n$ is the total number of observations
- $y_i$ is the actual output for the $i^{th}$ observation
- $\hat{y}_i$ is the predicted output for the $i^{th}$ observation
- $\sum_{i=1}^{n}$ denotes the sum over all observations

Now the goal of linear regression is to find the parameters $\theta$ that minimize this error function.

### Gradient Descent
Now that we established an error term, we aim to minimize it by adjusting our parameters $\theta$. This is commonly done by using one of the most fundamental algorithms in machine learning: **Gradient Descent**. It operates on the principle of iteratively adjusting the parameters $\theta$ in the direction of the negative gradient of the error function. The process is akin to descending a hill in the steepest direction until reaching the lowest point, hence the name "gradient descent".

For MSE, the gradient of the cost function with respect to the parameters $\theta$ can be computed easily, and the update rule becomes:
$$ \theta_{new} = \theta_{old} - \alpha \cdot \frac{d}{d\theta} MSE(\theta) $$

Where:
$\alpha$ is the learning rate, which determines the step size during each iteration
$\frac{d}{d\theta}MSE(\theta)$ gives the direction of steepest ascent, which we subtract to move in the opposite direction (i.e., steepest descent)



By iteratively adjusting the parameters in this way, we can converge towards the minimum of the cost function, thus finding the best parameters for our linear regression model.

*Hint: In the lecture, we used an absolute error term. In practice, the mean squared error term is more efficient, because it naturally penalizes high error values.

def gradient_descent(theta, x, y, alpha=0.01, iterations=1000):
    for i in range(iterations):
        # forward pass
        y_hat = theta[0] + theta[1] * x
        
        # calculate error (residuals)
        error = y_hat - y
        
        # calculate gradient
        n = len(y) # number of samples
        grad_theta0 = (2/n) * np.sum(error)
        grad_theta1 = (2/n) * np.sum(error * x)

#### Creating train and test dataset
Train/Test Split involves splitting the dataset into training and testing sets that are mutually exclusive. After which, you train with the training set and test with the testing set. 
This will provide a more accurate evaluation on out-of-sample accuracy because the testing dataset is not part of the dataset that have been used to train the model. Therefore, it gives us a better understanding of how well our model generalizes on new data.

This means that we know the outcome of each data point in the testing dataset, making it great to test with! Since this data has not been used to train the model, the model has no knowledge of the outcome of these data points. So, in essence, it is truly an out-of-sample testing.

Let's split our dataset into train and test sets. 80% of the entire dataset will be used for training and 20% for testing. We create a mask to select random rows using __np.random.rand()__ function: 


````python
# train test split
x_train, x_test, y_train, y_test = train_test_split(cdf[['ENGINESIZE']], cdf[['CO2EMISSIONS']], test_size=0.3, random_state=42)
````

#### Evaluation
We compare the actual values and predicted values to calculate the accuracy of a regression model. Evaluation metrics provide a key role in the development of a model, as it provides insight to areas that require improvement.

There are different model evaluation metrics: 
* Mean Absolute Error: It is the mean of the absolute value of the errors. This is the easiest of the metrics to understand since it’s just average error.

* Mean Squared Error (MSE): Mean Squared Error (MSE) is the mean of the squared error. It’s more popular than Mean Absolute Error because the focus is geared more towards large errors. This is due to the squared term exponentially increasing larger errors in comparison to smaller ones.

* Root Mean Squared Error (RMSE): This is the square root of the Mean Square Error.

* R-squared is not an error, but rather a popular metric to measure the performance of your regression model. It represents how close the data points are to the fitted regression line. The higher the R-squared value, the better the model fits your data. The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse).


````python
from sklearn import metrics

def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)

    MAE = metrics.mean_absolute_error(y_test, y_pred)
    MSE = metrics.mean_squared_error(y_test, y_pred)
    RMSE = np.sqrt(MSE)
    R2_Score = metrics.r2_score(y_test, y_pred)

    return pd.DataFrame([MAE, MSE, RMSE, R2_Score], index=['MAE', 'MSE', 'RMSE', 'R2-Score'], columns=[model_name])
````

````python
# Evaluate the model using the model evaluation function
evaluate_model(regr, x_test, y_test, "Simple Linear Regression")
````

<h2 id="multiple_regression_model">Multiple Regression Model</h2>

In reality, there are multiple variables that impact the co2emission. When more than one independent variable is present, the process is called multiple linear regression. As such we fit parameters $\theta=\theta_0,...,\theta_n-1$, with n being the number of independent variables, to multiplefeatures in our dataset. An example of multiple linear regression is predicting `CO2EMISSIONS` using the features `FUELCONSUMPTION_COMB` and `ENGINESIZE` of cars. Compared to Simple Linear Regression, implementation of Multiple Linear Regression is just as straightforward.

````python
# create new train test split
x_train, x_test, y_train, y_test = train_test_split(cdf[['ENGINESIZE','FUELCONSUMPTION_COMB']], cdf[['CO2EMISSIONS']], test_size=0.3, random_state=42)
````

````python
# fit new model
mult_regr_model = linear_model.LinearRegression()

mult_regr_model.fit(x_train, y_train)
````

````python
# print coefficients
print ('Coefficients: ', mult_regr_model.coef_)
print ('Intercept: ', mult_regr_model.intercept_)
````

````python
# evaluate the new model
evaluate_model(mult_regr_model, x_test, y_test, "Multiple Linear Regression")
````

````python
# Importing required libraries for 3D plotting
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

# Extract coefficients and intercept from the trained model
coefficients = mult_regr_model.coef_[0]
intercept = mult_regr_model.intercept_

print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")

# Create a grid for the regression surface
x1_range = np.linspace(x_test['ENGINESIZE'].min(), x_test['ENGINESIZE'].max(), 50)
x2_range = np.linspace(x_test['FUELCONSUMPTION_COMB'].min(), x_test['FUELCONSUMPTION_COMB'].max(), 50)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

# Calculate predicted CO2EMISSIONS for the grid
y_pred_grid = (
    coefficients[0] * x1_grid + coefficients[1] * x2_grid + intercept
)

# 3D Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of actual test data
ax.scatter(
    x_test['ENGINESIZE'], 
    x_test['FUELCONSUMPTION_COMB'], 
    y_test['CO2EMISSIONS'], 
    color='blue', 
    label='Actual Data', 
    alpha=0.7
)

# Plot regression surface
ax.plot_surface(
    x1_grid, x2_grid, y_pred_grid, 
    color='red', 
    alpha=0.5, 
    label='Regression Plane'
)

# Add labels and title
ax.set_xlabel('Engine Size')
ax.set_ylabel('Fuel Consumption (Combined)')
ax.set_zlabel('CO2 Emissions')
ax.set_title('3D Visualization of Multiple Linear Regression')
plt.legend(loc='upper left')
plt.show()
````
